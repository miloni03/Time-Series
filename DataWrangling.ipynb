```python
from pandasql import sqldf 

dfpsql = pd.read_csv(r'/content/daily-total-female-births-CA-with_nulls.csv') 

Query_string = """ select * from dfpsql limit 5 """ 

sqldf(Query_string, globals()) 
```

```python
#Applying a Filter: 
dfpsql[(dfpsql['births'] >=35) & (dfpsql['births'] <=40)] 
```

```python
Query_string = """ select * from dfpsql where births>=35 and births<=40 """ 

sqldf(Query_string, globals()) 
```

```python
#Distinct (Unique): 
dfpsql['births'].unique() 
```

```python
Query_string = """ select distinct births from dfpsql;""" 

sqldf(Query_string, globals()) 
```

```python
#IN: 
dfpsql[dfpsql.births.isin([20,30,40])] 
```

```python
Query_string = """ select * from dfpsql where births in(20,30,40);""" 

sqldf(Query_string, globals()) 
```

```python
#NOT IN
dfpsql[~dfpsql.births.isin([20,30,40])] 
```

```python
Query_string = """ select * from dfpsql where births not in(20,30,40);""" 

sqldf(Query_string, globals()) 
```

```python
#Ascending Data Order: 
dfpsql.sort_values(by = ['date','births'], ascending= True) 
```

```python
Query_string = """ select * from dfpsql order by date,births;""" 

sqldf(Query_string, globals()) 
```

```python
#Descending Data Order: 
dfpsql.sort_values(by = ['date','births'], ascending= False) 
```

```python
Query_string = """ select * from dfpsql order by date,births;""" 

sqldf(Query_string, globals()) 
```

```python
#Aggregation: 
dfpsql.agg({'births': ['count','min', 'max', 'mean']}) 
```

```python
Query_string = """ select count(births) as count, min(births) as min, max(births) as max, avg(births) as mean from dfpsql;""" 

sqldf(Query_string, globals()) 
```

```python
#GROUP BY: 
dfpsql.groupby('date')['births'].sum() 
```

```python
Query_string = """ select date , sum(births) as births from dfpsql group by date;""" 

sqldf(Query_string, globals()) 
```

```python
#GROUP BY with Aggregation: 
dfpsql.groupby('births').agg({'births': ['mean','min','max']}) 
```

```python
Query_string = """ select births , avg(births) as mean, min(births) as min, max(births) as max from dfpsql 

group by births;""" 

sqldf(Query_string, globals()) 

 
```

```python
#Join (Merge): 
import pandas as pd 

data1 = {'Empid': [1011, 1012, 1013, 1014, 1015], 'Name': ['John', 'Rahul', 'Rick', 'Morty', 'Tim'],  'Designation': ['Manager', 'Research Engineer', ' Research Engineer', 'VP', 'Delivery Manager'], 
'Date_of_joining': ['01-Jan-2000', '23-sep-2006', '11-Jan-2012', '21-Jan-1991', '12-Jan-1990']} 

Emp_df = pd.DataFrame(data1, columns = ['Empid', 'Name', 'Designation','Date_of_joining']) 

Emp_df 
```

```python
data2 = {'Empid': [1011, 1017, 1013, 1019, 1015], 
'Deptartment': ['Management', 'Research', 'Research', 'Management', 'Delivery'],  'Total_Experience': [18, 10, 10, 28, 22]} 

Dept_df = pd.DataFrame(data2, columns = ['Empid', 'Deptartment', 'Total_Experience']) 

Dept_df 
```

```python
#INNER JOIN:  
pd.merge(Emp_df, Dept_df, left_on='Empid',right_on='Empid', how='inner') 
```

```python
#LEFT JOIN: 
pd.merge(Emp_df, Dept_df,left_on='Empid',right_on='Empid', how='left') 
```

```python
#RIGHT JOIN: 
pd.merge(Emp_df, Dept_df,left_on='Empid',right_on='Empid', how='right') 
```

```python
#OUTER JOIN: 
pd.merge(Emp_df, Dept_df,left_on='Empid',right_on='Empid', how='outer') 
```

```python
#Resampling: 
import pandas as pd 

df = pd.read_csv(r'/content/daily-total-female-births-CA-with_nulls.csv',index_col =0, parse_dates=['date']) 

df.head(5) 
```

```python
#Resampling by Year: 
df.births.resample('Y').mean() 
```

```python
#Resampling by Week: 
df.births.resample('W').mean() 
```

```python
#Resampling on a Semimonthly basis: 
df.births.resample('SM').mean() 
```

```python
#Windowing Function: 
import pandas as pd 

dfExcelwin = pd.read_csv(r'/content/daily-total-female-births-CA-with_nulls.csv',index_col =0, parse_dates=['date']) 


dfExcelwin.head(5) 
```
